```python
#Library
import pandas as pd
import numpy as np
import skfuzzy as sf
import matplotlib.pyplot as plt
```

```python
#Himpunan fuzzy
r_dt = [500, 500, 1200] #gaji sedikit dalam dollar
r_sg = [1000, 2000, 3200] #gaji sedang
r_bk = [3000, 4000, 5000] #gaji banyak
```

```python
r_graph = np.arange(500, 5001)
r_graph
```

```python
fset = [r_dt, r_sg, r_bk]
lset = ['Sedikit', 'Sedang', 'Banyak']
```

```python
def drawMembershipTriangular(r_graph, fset, lset):
  deg = []
  for f in fset:
    deg.append(sf.trimf(r_graph, f))

  fig, ax = plt.subplots()
  for mf, label in zip(deg, lset):
    ax.plot(r_graph, mf, label=label)

  ax.legend()
  fig.show()
```

```python
drawMembershipTriangular(r_graph, fset, lset)
```
